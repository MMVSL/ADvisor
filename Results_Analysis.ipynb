## In this notebook we provide an example for analizing results obtained from AD search and application

```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
res1 = pd.read_csv("Out1.csv") #run Compare_AD_Strategies.py to obtain Out1.csv
best_approaches = res1[:5]
best_approaches
```

```python
for idx, row in best_approaches.iterrows():
    r2_in = float(row['r2_in'])
    r2_out = float(row['r2_out'])
    plt.figure(figsize=(4, 4))  
    plt.bar(['R2 In', 'R2 Out'], [r2_in, r2_out], color=['skyblue', 'salmon'])
    plt.ylim(0, 1) 
    plt.title(f'Approach {idx +1} - IN vs OUT performance')
    plt.ylabel('R2 score')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
res2 = pd.read_csv("Out2.csv") #run Calculate_AD.py to obtain Out2.csv
res2_in = res2[res2["AD_Label"]=="IN"]
res2_out = res2[res2["AD_Label"]=="OUT"]
print("Total number of IN compounds in query set: ", len(res2_in))
print("Total number of OUT compounds in query set: ", len(res2_out))
```
